```python

import matplotlib.pyplot as plt
import numpy as np
import os

trace_names = ["all_gather_0.txt", "all_gather_0.1.txt"]

for name in trace_names:
    trace = np.loadtxt(f"../{name}", dtype=np.float32)
    # sort
    trace = trace[trace[:,2].argsort()]
    plt.plot(trace[:,2], label=name)
plt.legend()
plt.show()
```

```python
for name in trace_names:
    trace = np.loadtxt(f"../{name}")
    # sort
    trace = trace[trace[:,2].argsort()]
    trace = np.cumsum(trace[:,0], axis=0)
    plt.plot(trace, label=name)
plt.legend()
plt.show()
```
